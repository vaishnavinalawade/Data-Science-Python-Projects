```python
act_minimum_wage = pd.DataFrame()
```

```python
for name, group in df.groupby("State"):
    if act_minimum_wage.empty:
        act_minimum_wage = group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name})
    else:
        act_minimum_wage = act_minimum_wage.join(group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name}))
```

```python
minimum_wage_corr = act_minimum_wage.replace(0,np.NaN).dropna(axis=1).corr().head(25)
```

```python
for problem in issue_df['State'].unique():
    if problem in minimum_wage_corr.columns:
        print("We're missiong some data!!")
```
